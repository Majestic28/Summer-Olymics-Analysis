# Summer Olympics Data Analysis Assignment

```python
import pandas as pd
import matplotlib.pyplot as plt
df = pd.read_csv("summer.csv")
df.head()
```

### 1. In how many cities Summer Olympics is held so far?

```python
print("In " + str(len(df['City'].unique())) + " cities Summer Olymics is held so far")
```

### 2. Which sport is having most number of Gold Medals so far? (Top 5)

```python
t = df[df['Medal'] == 'Gold']

lt = []
for Sport in t['Sport'].unique():
    lt.append([Sport, len(t[t['Sport'] == Sport])])
lt
```

```python
print("In " + str(len(t['Sport'].unique())) + " sports having most number of Gold Medals so far!!")
print("Total Golds: "+ str(len(t['Sport'])))
```

```python
lt = pd.DataFrame(lt, columns = ['Sport', 'Gold Count'])
lt = lt.sort_values(by = 'Gold Count', ascending=False)
lt = lt.head()
lt.plot(x ='Sport',y='Gold Count',kind = 'bar', figsize = (10,5))
```

### 3. Which sport is having most number of medals so far? (Top 5)

```python
ot = []
dat = df.groupby('Sport')
for sport in dat:
  ot.append([sport[0],len(sport[1])])
ot
```

```python
ot = pd.DataFrame(ot, columns = ['Sport', 'Medals Count'])
ot = ot.sort_values(by = 'Medals Count', ascending=False)
ot = ot.head()
ot.plot(x ='Sport',y='Medals Count',kind = 'bar', figsize = (10,5))
```

### 4. Which player has won most number of medals? (Top 5)

```python
lst = []
data = df.groupby('Athlete')
for sport in data:
  lst.append([sport[0],len(sport[1])])
lst
```

```python
lst = pd.DataFrame(lst, columns = ['Athlete', 'Medals Count'])
lst = lst.sort_values(by = 'Medals Count', ascending=False)
lst = lst.head()
lst.plot(x ='Athlete',y='Medals Count',kind = 'bar', figsize = (10,5))
```

### 5. Which player has won most number Gold Medals of medals? (Top 5)

```python
value = df[df['Medal'] == "Gold"]
lst = []
for Athlete in value['Athlete'].unique():
    lst.append([Athlete, len(value[value['Athlete'] == Athlete])])
    #lst.append(len(value[value['Athlete'] == Athlete]))
lst
```

```python
lst = pd.DataFrame(lst, columns = ['Athlete', 'Gold Count'])
lst = lst.sort_values(by = 'Gold Count', ascending=False)
lst = lst.head()
lst.plot(x ='Athlete',y='Gold Count',kind = 'bar', figsize = (10,5))
```

### 6. In which year India won first Gold Medal in Summer Olympics?

```python
df1 = df[df['Country'] == 'IND']
df1 = df1[df1["Medal"] == "Gold"]
print("India won Gold Medal in the year " + str(df1['Year'].min()) + " of Summer Olympics")
```

```python
lst = []
for year in df1['Year'].unique():
    lst.append([year, len(df1[df1['Year'] == year])])
lst
```

```python
lst = pd.DataFrame(lst, columns = ['Year', 'Gold Count'])
lst = lst.sort_values(by = 'Year', ascending=True)
lst.plot(x ='Year',y='Gold Count',kind = 'bar', figsize = (10,5))
```

### 7. Which event is most popular in terms on number of players? (Top 5)

```python
count = []
items = df['Event'].unique()
for item in items:
  count.append([item,len(df[df['Event'] == item])])
count
```

```python
count = pd.DataFrame(count, columns = ['Event', 'No of Players'])
count = count.sort_values(by = 'No of Players', ascending=False)
count = count.head()
count.plot(x ='Event',y='No of Players',kind = 'bar', figsize = (10,5))
```

### 8. Which sport is having most female Gold Medalists? (Top 5)

```python
data = df[df['Gender'] == 'Women']
data = data[data['Medal'] == "Gold"]
print("The total gold medal got by females are " + str(len(data['Sport'])))
```

```python
lst = []
for Sport in data['Sport'].unique():
    lst.append([Sport, len(data[data['Sport'] == Sport])])
lst
```

```python
lst = pd.DataFrame(lst, columns = ['Sport', 'Count'])
lst = lst.sort_values(by = 'Count', ascending=False)
lst = lst.head()
lst.plot(x ='Sport',y='Count',kind = 'bar', figsize = (10,5))
```
